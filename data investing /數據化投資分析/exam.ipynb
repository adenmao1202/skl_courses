```python
import yfinance as yf
import numpy as np
import pandas as pd
from scipy.optimize import minimize

# Define the tickers
tickers = ['2330.TW', '2317.TW', '2454.TW', '2881.TW', '2382.TW',
           '2308.TW', '2882.TW', '2412.TW', '2891.TW', '3711.TW']

# Download monthly adjusted close prices for the full sample period
data = yf.download(tickers=tickers, interval='1mo', start='2019-01-01', end='2023-12-31')['Adj Close']

# Drop any rows with missing data
data.dropna(inplace=True)

# Since data is already beginning-of-month prices, we use it directly

# Calculate monthly returns using the formula: Returns = (P[t] - P[t-1]) / P[t-1]
returns = data.pct_change().dropna()

# Calculate the covariance matrix
cov_matrix = returns.cov()

# Number of assets
num_assets = len(tickers)

# Function to calculate portfolio variance
def portfolio_variance(weights):
    return np.dot(weights.T, np.dot(cov_matrix, weights))

# Constraints: Sum of weights equals 1
constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})

# Bounds: No short selling allowed, weights between 0 and 1
bounds = tuple((0, 1) for _ in range(num_assets))

# Initial guess (equal allocation)
initial_guess = num_assets * [1.0 / num_assets]

# Minimize portfolio variance
opt_result = minimize(portfolio_variance, initial_guess,
                      method='SLSQP', bounds=bounds, constraints=constraints,
                      options={'ftol': 1e-20, 'disp': True})

# Extract the optimal weights
gmvp_weights = opt_result.x

# Find the weight on 2881.TW
ticker_index = tickers.index('2881.TW')
weight_2881 = gmvp_weights[ticker_index]

# Report the weight with at least 3 decimal points
print(f"Weight on 2881.TW in the constrained GMVP (no short selling): {weight_2881:.3f}")
```
