```python
import yfinance as yf
import pandas as pd

df = yf.download(['TSLA', 'GOOGL', 'AAPL'],
                 interval="1d",
                
                 start="2024-09-01",
                 end="2024-10-01")
prc = df['Adj Close']
prc
```

```python
ret = prc.pct_change()
ret.head()
```

```python
ret.mean()*100
```

```python
ret.std()
```

```python
ret.cov()
```
